## Run the WebUI

Wait for the line "Running on local URL:  http://0.0.0.0:7860 " and click the link ending with .pinggy.link between the smiling emojis.  
It can take about a minute for the line to appear.

```python
%cd /kaggle/working/InstantMesh
       
from multiprocessing import Process
import sys
import time

open('log.txt', 'w').close()

def run_app():
    !python app.py & ssh -o StrictHostKeyChecking=no -p 80 -R0:localhost:7860 a.pinggy.io > log.txt
    
def print_url():
    print("waiting for output")
    time.sleep(2)
    sys.stdout.flush()
    
    found = False
    with open('log.txt', 'r') as file:
        end_word = '.pinggy.link'
        for line in file:
            #print(line)
            start_index = line.find("http:")
            if start_index != -1:
                end_index = line.find(end_word, start_index)
                if end_index != -1:
                    print("游때 游때 游때")
                    print("URL: " + line[start_index:end_index + len(end_word)])
                    print("游때 游때 游때")
                    found = True
    if not found:
        print_url()
    else:
        with open('log.txt', 'r') as file:
            for line in file:
                print(line)
    
p_app = Process(target=run_app)
p_url = Process(target=print_url)
p_app.start()
p_url.start()
p_app.join()
p_url.join()
    
```

## Start FileBrowser

* Follow the .pinggy.link link to interact with FileBrowser.  
* Outputs are in the folder working -> InstantMesh -> outputs  
* If the app starts with a login screen, the username and password are both "admin"

```python
%cd /kaggle/working/

!chmod a+x /kaggle/working/InstantMesh/filebrowser
!/kaggle/working/InstantMesh/filebrowser config set --auth.method=noauth > /dev/null
        
from multiprocessing import Process
import sys
import time

open('log.txt', 'w').close()

def run_app():
    !/kaggle/working/InstantMesh/filebrowser -c "/kaggle/working/config.json" & ssh -o StrictHostKeyChecking=no -p 80 -R0:localhost:8080 a.pinggy.io > log.txt 
    
def print_url():
    print("waiting for output")
    time.sleep(2)
    sys.stdout.flush()
    
    found = False
    with open('log.txt', 'r') as file:
        end_word = '.pinggy.link'
        for line in file:
            #print(line)
            start_index = line.find("http:")
            if start_index != -1:
                end_index = line.find(end_word, start_index)
                if end_index != -1:
                    print("游때 游때 游때")
                    print("URL: " + line[start_index:end_index + len(end_word)])
                    print("游때 游때 游때")
                    found = True
    if not found:
        print_url()
    else:
        with open('log.txt', 'r') as file:
            for line in file:
                print(line)
    
p_app = Process(target=run_app)
p_url = Process(target=print_url)
p_app.start()
p_url.start()
p_app.join()
p_url.join()
```
