## Installation

```python
# Install InstantMesh
%cd /kaggle/working
!git clone https://github.com/WandaWeb/InstantMesh
!pip install triton
%cd InstantMesh
!pip install -r requirements.txt

# Install FileBrowser
!wget https://github.com/filebrowser/filebrowser/releases/download/v2.27.0/linux-amd64-filebrowser.tar.gz
!tar xvfz linux-amd64-filebrowser.tar.gz
!chmod a+x filebrowser
!rm /kaggle/working/config.json
!/kaggle/working/InstantMesh/filebrowser config init 
!/kaggle/working/InstantMesh/filebrowser config set --auth.method=noauth > /dev/null
!/kaggle/working/InstantMesh/filebrowser config set --branding.theme=dark > /dev/null
!/kaggle/working/InstantMesh/filebrowser users add admin admin > /dev/null

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```
